```python
covid_df['Active_Cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
#Pivot Table
state=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"],index="State/UnionTerritory",aggfunc=max)
```

```python
#recovery rate

state["Recovery Rate"]=state["Cured"]*100/state["Confirmed"]
```

```python
#mortality rate

state["Mortality Rate"]=state["Deaths"]*100/state["Confirmed"]
```

```python
#sort (Based on confirm cases)

state.Confirmed.sort_values(ascending=False)
```

```python
#Top 10 active cases states

Top_10_Active=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
```

```python
ax=sns.barplot(data=Top_10_Active.iloc[:10],x="State/UnionTerritory",y="Active_Cases")
```

```python
#Top 10 active cases states

Top_10_Active=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
figure=plt.figure(figsize=(25,10))

plt.title("TOP 10 STATES WITH MOST ACTIVE CASES IN INDIA")
ax=sns.barplot(data=Top_10_Active.iloc[:10],x="State/UnionTerritory",y="Active_Cases")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show
```

```python
#top_10 states with highest deaths
Top_10_Death=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(20,10))
```

```python
#top_10 states with highest deaths
Top_10_Death=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(20,8))
plt.title("TOP 10 States with highest Number of deaths",size=14)
bx=sns.barplot(data=Top_10_Death.iloc[:12],y="Deaths",x="State/UnionTerritory")
plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```

```python
#Growth_trend
fig=plt.figure(figsize=(12,6))
ax= sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])])
ax.set_title("Top 5 Affected States in India",size=16)
plt.show()
```

```python
vaccination=vaccine_df.drop(columns=['AEFI','18-44 Years (Doses Administered)','Sputnik V (Doses Administered)',
                            '45-60 Years (Doses Administered)','60+ Years (Doses Administered)'],axis =1)
```

```python
vaccination.head()
```

```python
#Data descripancy removing

vaccine=vaccine_df[vaccine_df.State!='India']
```

```python
#Male vs Female vaccination

male=vaccination["Male(Individuals Vaccinated)"].sum()

female=vaccination["Female(Individuals Vaccinated)"].sum()

px.pie(names=["M","F"],values=[male,female],title ="Male and Female vaccination")
#plt.show()
```

```python
vaccine.head()
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated": "Total"},inplace=True)
```

```python
#most vaccinated state
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)
max_vac
```

```python
figure=plt.figure(figsize=(40,30))
plt.title("TOP 5 vaccinated sates in INDIA")
ax=sns.barplot(data=max_vac.iloc[:10],x=max_vac.index,y=max_vac.Total)
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
